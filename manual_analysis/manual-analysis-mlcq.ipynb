```python
import pandas as pd

mlcq = pd.read_csv(filepath_or_buffer="../data/smells/mlcq-official.csv", sep=";").set_index(["commit_hash", "code_name"]).sort_index()

mlcq_classes = mlcq[mlcq["type"] == "class"]
mlcq_classes
```

```python
pmd = pd.read_csv(filepath_or_buffer="../data/smells/complete_pmd.csv", sep=",")
pmd["entity"] = pmd["entity"].transform(lambda x: x.replace("$", "."))

pmd = pmd.set_index(["revision", "entity"]).sort_index()
pmd
```

```python
javametrics = pd.read_csv(filepath_or_buffer="../data/smells/complete_javametrics.csv", sep=",").set_index(["revision","entity"]).sort_index()

javametrics
```

```python
metrics = pd.merge(javametrics, pmd, on=["revision", "entity"], how="outer")

metrics
```

```python
all_smells_with_metrics = mlcq_classes.join(metrics, on=["commit_hash", "code_name"], how="left")
all_smells_with_metrics
```

```python
complete_smells_with_metrics = mlcq_classes.join(metrics, on=["commit_hash", "code_name"], how="inner")
complete_smells_with_metrics
```

```python
missing_metrics = pd.concat([complete_smells_with_metrics, all_smells_with_metrics]).drop_duplicates(keep=False)
missing_metrics
```
